```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv(r"C:\Users\oguzhanarar\Desktop\veri2.csv")
```

```python
model_zinciri = Pipeline([
    ('scaler', StandardScaler()),                 
    ('poly', PolynomialFeatures(degree=2)),      
    ('regressor', LinearRegression())            
])
```

```python
df = df.dropna()
```

```python
kolon = ["yıl","gün","saat","bt","bz","yogunluk","hız","kp"]
df.columns = kolon
```

```python
df = df.drop(["yıl","gün","saat"],axis=1)
```

```python
df = df[~(df["hız"]>990)]
```

```python
df = df[~(df["bz"]==0)]
```

```python
df[df["kp"]>80]
```

```python
X = df[["bt","bz","yogunluk","hız"]]
y = df["kp"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=15)
```

```python
oversample = SMOTE(k_neighbors=2)
```

```python
X_train_res, y_train_res = oversample.fit_resample(X_train, y_train)
```

```python
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train_res)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
scaler = StandardScaler()
X_train_final = scaler.fit_transform(X_train_poly)
X_test_final = scaler.transform(X_test_poly)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train_final, y_train_res)
```

```python
y_train_pred = model.predict(X_train_final)
y_test_pred = model.predict(X_test_final)
```

```python
r2_score(y_test/10, y_test_pred/10)
```
